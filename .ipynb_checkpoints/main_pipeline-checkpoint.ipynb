### 0.  Car Images -> Feature Extraction

To be able to find a car in an image, it's neccesary to have some idea of what we're looking for, we need to have something that defines the concept of a "car", this is a "model" of a car to compare later.

We can extract information from an image because the pixels from it represent "features" of information, this information will be useful.

### 1. Features 

#### Spatial Features

The simplest and useful, means a resizing of the image and flatteing with ravel(). 
Spatial features represents the values and ordering of the pixels.

All happens in the method called get_bin_spatial_features(), this method used an RGB image with a size 16x16

#### Color Histogram Features

The color histogram features identify the amount of certain color values.
Using other color spaces like HSV will make more easy detect cars in an image (RGB doesn't work so well), identify a car using HSV is much more useful and allow us to read a lot of color edges.

A method called get_color_hist_features() do this work, the parameters are the numbers of bins.

#### Hog Features

One of the most usefult features and the most expensive (computationally). HOG "histogram of oriented gradients" and basically be though of as colors are changing by 'p' amount in 'x' direction.HOG provides a generic shape information, the method called get_hog_features() do this work. Parameters to tune: pixels per cell, number of orientation directions, cells per block.
* Setup: 8 orientation directions (less data and the same performance of 9), 8 pixels per cell, 8 cell per block, all this parameters fit so well with de 64x64 training images.


```python
###########################################################################################
# Initial imports
###########################################################################################
import numpy as np
import cv2
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from skimage.feature import hog
from scipy.ndimage.measurements import label
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import glob
import time
import random

#import imageio 
#imageio.plugins.ffmpeg.download()
from moviepy.editor import VideoFileClip
from IPython.display import HTML

%matplotlib inline
print('Import done!')
```

```python
###########################################################################################
# Plot helpers
###########################################################################################    

# draw one big image (could fail because an unknown bug)
def one_plot(img):
    # Plots one big image    
    fig, ax = plt.subplots(figsize=(20, 10))
    ax.imshow(img)

# draw two medium images
def two_plots(img1, img2, img1_title=None, img2_title=None):
    # Plots two images
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
    f.tight_layout()
    ax1.imshow(img1)
    if img1_title: ax1.set_title(img1_title, fontsize=30)
    ax2.imshow(img2)
    if img2_title: ax2.set_title(img2_title, fontsize=30)
    plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)

print('Plot helpers done!')
```

```python
###########################################################################################
# 1 -> Features
###########################################################################################    

# return binned color features for an image
def get_bin_spatial_features(img, size=(16, 16)):
    # Resize the image, and then flatten it out into a feature vector
    return cv2.resize(img, size).ravel()

# return color histogram feature
def get_color_hist_features(img, nbins=32, bins_range=(0, 256)):
    # calculate the histogram of the color channels separately
    ch1_hist = np.histogram(img[:,:,0], bins=nbins, range=bins_range)[0]
    ch2_hist = np.histogram(img[:,:,1], bins=nbins, range=bins_range)[0]
    ch3_hist = np.histogram(img[:,:,2], bins=nbins, range=bins_range)[0]
    # concatenate the histograms into a single feature vector
    return np.concatenate((ch1_hist, ch2_hist, ch3_hist))

# return HOG features
def get_hog_features(img, orient, pix_per_cell, cell_per_block, vis=False, feature_vec=True):
    if vis == True:
        features, hog_image = hog(img, orientations=orient, 
                                  pixels_per_cell=(pix_per_cell, pix_per_cell),
                                  cells_per_block=(cell_per_block, cell_per_block), 
                                  transform_sqrt=True, 
                                  visualise=vis, feature_vector=feature_vec)
        return features, hog_image
    else:      
        features = hog(img, orientations=orient, 
                       pixels_per_cell=(pix_per_cell, pix_per_cell),
                       cells_per_block=(cell_per_block, cell_per_block), 
                       transform_sqrt=True, 
                       visualise=vis, feature_vector=feature_vec)
        return features

# take a list of images files names and put it into a list of image features, using later into the classifier
def extract_features(imgs, color_space='HSV', spatial_size=(16,16), hist_bins=32,
                     orient=8, pix_per_cell=8, cell_per_block=2, hog_channel='ALL', precomputed_hog=None,
                     use_spatial_features=True, use_color_hist_features=True, use_hog_features=True):    
    
    features = []
    for img in imgs:
        img_features = []
        if type(img) is str:
            image = mpimg.imread(img)
            if img.endswith('.png'):
                image = image*255
        else:
            image = img
            
        img_size = (64, 64)
        if image.shape[:2] != img_size:
            image = cv2.resize(image, img_size)
        if precomputed_hog is not None and precomputed_hog.shape[:2] != img_size:
            precomputed_hog = cv2.resize(precomputed_hog, img_size)
        
        # a basic color conversion
        original_image = np.copy(image)
        if color_space == 'RGB':
            image = np.copy(image)
        else:
            color = eval('cv2.COLOR_RGB2{}'.format(color_space))
            image = cv2.cvtColor(image, color)

        # spatial features
        if use_spatial_features == True:
            spatial_features = get_bin_spatial_features(original_image, size=spatial_size)
            img_features.append(spatial_features)
            
        # color histogram Features
        if use_color_hist_features == True:
            hist_features = get_color_hist_features(image, nbins=hist_bins)/255
            img_features.append(hist_features)
            
        # HOG Features
        if use_hog_features == True:
            if precomputed_hog is not None:
                img_features.append(precomputed_hog*255)
            else:
                if hog_channel == 'ALL':
                    hog_features = []
                    for channel in range(image.shape[2]):
                        hog_features.append(get_hog_features(
                            image[:,:,channel], 
                            orient,
                            pix_per_cell,
                            cell_per_block, 
                            vis=False,
                            feature_vec=True
                        ))
                    hog_features = np.ravel(hog_features)        
                else:
                    hog_features = get_hog_features(
                        image[:,:,hog_channel],
                        orient, 
                        pix_per_cell,
                        cell_per_block,
                        vis=False,
                        feature_vec=True
                    )

                img_features.append(hog_features*255)
            
        # add this to de principal list
        features.append(np.concatenate(img_features))
    if len(imgs) == 1:
        return np.array(features)
    
    return features
print('Features done!')
```

### 2. Training a classifier

Now that we can get features from images, we can use this features to train a ML classifier for vehicle detection.
17760 images where used (images from cars (~50%) and images from non-cars (~50%). These features are labeled as "0" (is a car) and "1" (non a car), later passed into a SVM, using SVC. (linearly separates the data).

Here are links to the labeled data for vehicles https://s3.amazonaws.com/udacity-sdc/Vehicle_Tracking/vehicles.zip and non-vehicles https://s3.amazonaws.com/udacity-sdc/Vehicle_Tracking/non-vehicles.zip

Feature values are normalized before being passed in to improve learning. The data was shuffle.

The final test accuracy was ~ 99.2% and took ~ 100 secs.

At the end examples of original image, raw features, normalized features and hog-visualization are plotted.

```python
###########################################################################################
# 2 -> Training a classifier
###########################################################################################    

vehicle_imgs = shuffle(glob.glob('./training-data/vehicles/*/*.png'))
nonvehicle_imgs = shuffle(glob.glob('./training-data/non-vehicles/*/*.png'))

n_imgs = len(vehicle_imgs) + len(nonvehicle_imgs)
# take the time
t = time.time()
vehicle_features = extract_features(vehicle_imgs)
nonvehicle_features = extract_features(nonvehicle_imgs)
t2 = time.time()
print('Total extraction time:', round(t2-t, 2), 'secs,', n_imgs, 'image features')

x = np.vstack((vehicle_features, nonvehicle_features)).astype(np.float64)
scaler = StandardScaler().fit(x)
# normalize the features
scaled_x = scaler.transform(x)  
# labels vector - 1 is car, 0 is not
y = np.hstack((np.ones(len(vehicle_features)), np.zeros(len(nonvehicle_features))))

min_scaled = np.min(scaled_x)
max_scaled = np.max(scaled_x)

# shuffle data train/test split
rand = random.randint(0, 100)
x_train, x_test, y_train, y_test = train_test_split(scaled_x, y, test_size=0.25, random_state=rand)

# svc training
print('Training the SVC...')
svc = LinearSVC()
t = time.time()  # Track the training time
svc.fit(x_train, y_train)
t2 = time.time()
print('Training time:', round(t2-t, 2), 'secs')

# svc accuracy
print('Accuracy of the SVC:', round(svc.score(x_test, y_test), 3))

# Plot an example of raw and scaled features
car_ind = random.randint(0, len(vehicle_imgs))
fig = plt.figure(figsize=(12,4))
plt.subplot(131)
plt.imshow(mpimg.imread(vehicle_imgs[car_ind]))
plt.title('Original Image')
plt.subplot(132)
plt.plot(x[car_ind])
plt.title('Raw Features')
plt.subplot(133)
plt.plot(scaled_x[car_ind])
plt.title('Normalized Features')
fig.tight_layout()

ind = car_ind
# Read in the image
image = mpimg.imread(vehicle_imgs[car_ind])
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
# Define HOG parameters
orient = 9
pix_per_cell = 8
cell_per_block = 2
# Call our function with vis=True to see an image output
features, hog_image = get_hog_features(gray, orient, 
                        pix_per_cell, cell_per_block, 
                        vis=True, feature_vec=False)
# Plot the examples
fig = plt.figure()
plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.title('Example Car Image')
plt.subplot(122)
plt.imshow(hog_image, cmap='gray')
plt.title('HOG Visualization')

print('Features && SVM done! car_ind =', car_ind)
```

### 3. Finding cars, sliding windows

The SVM was trained on 64x64 pixel images, it's neccesary to find a way to feed it similar input if we want keep the accuracy (~99.2%). 

The search was developed using sliding windows. Windows of various sizes will be passed over parts of the image to capture many sub-images, each of these sub-images will then be tested with the classifier to tell us whether the sub-image contains a car or not.

In the output you see the many windows drawn onto the image. A high 75% overlap is used to avoid missing centered images of the vehicles. To reduce computation time, I only search the right half of the image because the pictures are taken from a vehicle in the far-left lane.

```python
###########################################################################################
# 3 -> Finding cars, sliding windows
###########################################################################################    
# returns the list of windows to search through
# sizes = the square dimension of the windows
# y_start_stop, x_start_stop = percent of the image where start and stop in each axis
def get_windows(img, sizes=[128, 180, 256], overlap=.75, y_start_stop=(.5, .92), x_start_stop=(.5, 1)): 
    y_size, x_size = img.shape[:2]
    x_start = int(x_start_stop[0]*x_size)
    x_stop = int(x_start_stop[1]*x_size)
    y_start = int(y_start_stop[0]*y_size)
    y_stop = int(y_start_stop[1]*y_size)
    
    # windows will be stored in this array
    windows = [] 
    for size in sizes:
        step = int(size*(1-overlap))
        for y in range(y_start, y_stop-size, step):
            for x in range(x_start, x_stop-size, step):  # windows from the left
                windows.append(((x, y), (x+size, y+size)))
            for x in range(x_stop, x_start+size, -step):  # windows from the right
                windows.append(((x-size, y), (x, y+size)))           
    return windows
    
# bounding boxes
def draw_boxes(img, boxes, color=(0, 0, 255), thickness=4):
    imcopy = np.copy(img)
    for box in boxes:
        cv2.rectangle(imcopy, box[0], box[1], color, thickness)  
    return imcopy

# test 1
image = mpimg.imread('./test_images/test4.jpg')
draw_img = np.copy(image)
# only for demonstration (different windows sizes)
boxes_1 = get_windows(image, sizes=[128], y_start_stop=(.4, .7), x_start_stop=(.1, 1))
boxes_2 = get_windows(image, sizes=[180], y_start_stop=(.4, 1), x_start_stop=(.1, 1))
boxes_3 = get_windows(image, sizes=[256], y_start_stop=(.4, 1), x_start_stop=(.1, 1))

draw_img = draw_boxes(draw_img, boxes_1, color=(0, 255, 0), thickness=3)
draw_img = draw_boxes(draw_img, boxes_2, color=(0, 0, 255), thickness=2)
draw_img = draw_boxes(draw_img, boxes_3, color=(255, 0, 0), thickness=1)
windows = np.concatenate((boxes_1, boxes_2, boxes_3))
print(len(windows), 'windows test 1')
one_plot(draw_img)

# test 2
image = mpimg.imread('./test_images/test4.jpg')
draw_img = np.copy(image)
# only for demonstration (different windows sizes)
boxes_1 = get_windows(image, sizes=[128], y_start_stop=(.5, .7))
boxes_2 = get_windows(image, sizes=[120])
boxes_3 = get_windows(image, sizes=[170])

draw_img = draw_boxes(draw_img, boxes_1, color=(0, 255, 0), thickness=3)
draw_img = draw_boxes(draw_img, boxes_2, color=(0, 0, 255), thickness=2)
draw_img = draw_boxes(draw_img, boxes_3, color=(255, 0, 0), thickness=1)
windows = np.concatenate((boxes_1, boxes_2, boxes_3))
print(len(windows), 'windows test 1')
one_plot(draw_img)
```

### 4. Boxes around cars, heatmaps

The sliding-window results in > 500 windows, too much! Each of these windows, these sub-images, will be tested by extracting features and passing those features to the classifier using the contains_car method. 

To identify a single bounding-box per car a heatmap is used, where windows overlap the regions gets hotter. A heatmap identifies high concentrations of windows. Using the heatmap information allows to draw a single bounding-box around each car. The false positives can be avoided by summing the heatmaps across multiple frames (of a video) and taking a higher threshold. get_aggregate_heatmap along with the stored prev_heatmaps do this work.

```python
###########################################################################################
# 4 -> Boxes around cars, heatmaps
###########################################################################################   
# true or false for whether the window of the given image contains a car
def contains_car(img, window):
    top_left, bottom_right = window[0], window[1]
    sub_img = img[top_left[1]:bottom_right[1], top_left[0]:bottom_right[0]]
    features = extract_features([sub_img])
    # normalize the features
    scaled_features = scaler.transform(features)  
    # was getting a single HUGE value, so clipping
    scaled_features = np.clip(scaled_features, min_scaled, max_scaled)  
    return svc.predict(scaled_features) == 1
    
def get_heatmap(img, boxes, thresholded=False):
    heatmap = np.zeros_like(img[:,:,0])
    for box in boxes:
        top_left, bottom_right = box[0], box[1]
        heatmap[top_left[1]:bottom_right[1], top_left[0]:bottom_right[0]] += 1
    if thresholded: return threshold_heatmap(heatmap)
    return heatmap

def threshold_heatmap(heatmap, threshold=2):
    heatmap[heatmap < threshold] = 0
    return heatmap

def get_aggregate_heatmap(thresholded=True, threshold=4):
    global prev_heatmaps
    heatmaps = np.array(prev_heatmaps)
    if not thresholded: return np.sum(heatmaps, axis=0)
    return threshold_heatmap(np.sum(heatmaps, axis=0), threshold=threshold)

def draw_heatmap_boxes(img, heatmap):
    label_img, n_items = label(heatmap)
    # iterate through all detected cars
    for car_number in range(1, n_items+1):
        nonzero = (label_img == car_number).nonzero()  # find pixels with each car_number label value
        nonzeroy = np.array(nonzero[0])
        nonzerox = np.array(nonzero[1])
        # bounding box
        box = ((np.min(nonzerox), np.min(nonzeroy)), (np.max(nonzerox), np.max(nonzeroy)))
        cv2.rectangle(img, box[0], box[1], (0, 0, 255), 6)
    return img

# tests
image = mpimg.imread('./test_images/test4.jpg')
draw_img = np.copy(image)

windows_1 = get_windows(image, sizes=[80], overlap=.75, y_start_stop=(.5, .7))
windows_2 = get_windows(image, sizes=[120, 180], overlap=.75)
windows = windows_1 + windows_2
print(len(windows), 'windows')

car_windows = []
for window in windows:
    if contains_car(image, window):
        car_windows.append(window)
        
draw_img = draw_boxes(draw_img, car_windows)
heatmap = get_heatmap(image, car_windows, thresholded=False)
heatmap_boxes = draw_heatmap_boxes(np.copy(image), heatmap)
two_plots(draw_img, heatmap_boxes)
cv2.imwrite('./output_images/windows.jpg', cv2.cvtColor(draw_img, cv2.COLOR_RGB2BGR))
cv2.imwrite('./output_images/cars.jpg', cv2.cvtColor(heatmap_boxes, cv2.COLOR_RGB2BGR))
```

### 5. Basic pipeline

All the steps grouped into a simple pipeline function, find_cars(), given an image the function will return the same imgae with bounding boxes around to detected cars.

```python
###########################################################################################
# 5 -> Basic pipeline
###########################################################################################   
prev_heatmaps = []
def find_cars(image, vis=False):
    global prev_heatmaps
    
    windows_1 = get_windows(image, sizes=[80], y_start_stop=(.4, .7))
    windows_2 = get_windows(image, sizes=[120, 180])
    windows = windows_1 + windows_2
    
    car_windows = []
    for window in windows:
        if contains_car(image, window):
            car_windows.append(window)
            
    heatmap = get_heatmap(image, car_windows, thresholded=False)     
    prev_heatmaps.insert(0, heatmap)
    if len(prev_heatmaps) > 5:
        prev_heatmaps.pop()
    aggregate_heatmap = get_aggregate_heatmap(threshold=4) 
    if vis == True:
        two_plots(heatmap, aggregate_heatmap, 'Heatmap', 'Agg.Heatmap')
    return draw_heatmap_boxes(np.copy(image), aggregate_heatmap)

image = mpimg.imread('./test_images/test4.jpg')
one_plot(find_cars(image, True))
```

### 6. Processing videos
It's time to test the pipeline with videos, each frame is passed through the pipeline, using heatmaps and aggregated_heatmaps for the last 5 frames and summed together to avoid false-positives. Finally using all this information, each frame will have the bounding-box over each detected car.

The pipeline produce a new video with the bounding-boxes over each detected car.

The output videos are stored in the directory "output_videos".

```python
###########################################################################################
# 6 -> Processing videos
###########################################################################################   

# pipeline of test_video.mp4
prev_heatmaps = []
file_name = 'test_video.mp4'
video = VideoFileClip('./test_videos/{}'.format(file_name))  # get the input video
video = video.fl_image(find_cars)  # pipe video into the find_cars pipeline
%time video.write_videofile('./output_videos/{}'.format(file_name), audio=False)  # save the new video
```

### 7. Conclusions && results

The final pipeline has one big problem, the computation time, is too slowly, for the project_video.mp4 with 50 second of duration it took ~ 1:21 minutes to process!!!, for a real time application this wouldn't work.
I guess that te bottleneck is feature-extraction, all the feature-set need to be extracted in each frame of the ~500 sliding windows. HOG is the most expensive feature. Other important factor in the computation time is the number of sliding windows where the search is performed, that produces a big problem on performance.
In the final output video there are some false positives, maybe with a better tunning of heatmaps, this could be solved.

Perhaps exist a better method to detect and track cars on video, replacing the HOG + SVM pipeline. I found a paper on https://arxiv.org/pdf/1512.02325.pdf using a deep network to perform the detection.
